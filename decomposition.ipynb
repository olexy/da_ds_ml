```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
from statsmodels.tsa.seasonal import seasonal_decompose

%matplotlib inline 

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/AirPassengers.csv')
df
```

```python
df.set_index('Month',inplace=True)
df.index=pd.to_datetime(df.index)
#drop null values
df.dropna(inplace=True)
df.plot()
```

```python
result=seasonal_decompose(df['#Passengers'], model='multiplicable') #Use multiplicable model because the variance is non-linear
```

```python
result.seasonal.plot()
```

```python
result.trend.plot()
```

```python
result.resid.plot()
```

```python
np.log(result.resid).plot()
```
