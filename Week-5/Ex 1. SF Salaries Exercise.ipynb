** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
job_titles = sal['JobTitle'].str.upper().values
job_titles
```

```python
sum([True for title in job_titles if 'CHIEF' in title])
```

```python
len([True for title in job_titles if 'CHIEF' in title])
```

```python
len([True for title in sal['JobTitle'].str.upper() if 'CHIEF' in title])
```
