```python
np.random.seed(101)
x = np.linspace(0.1,3, 20).round(2)
y = np.round(np.sin(x) + 0.1*np.random.randn(20), 2)
p = np.poly1d(np.polyfit(x,y,8))
print(p), p([0, np.pi])
```

```python
fig = plt.figure(figsize = (8,5))  # this is to create a plot Axis (siize is provided as tuple)

plt.plot(x,y, 'bo', label='Original Data')

t = np.linspace(0.1, 3, 100)
plt.plot(t, p(t), 'r-*', label='Poly Fit')

plt.plot(t, np.sin(t), label='True function')

plt.xlabel('X values 0 to $\pi$', fontsize=18)
plt.ylabel('Y values ', fontsize=18)
plt.title('Curve Fitting', fontsize=21)
plt.legend(fontsize=15, loc = 'lower center')

# plt.xticks(xticks)
plt.grid(True)
plt.show()

fig.savefig('name of the figure.png', dpi=200)
```

```python
# you can use MSE or R2  >> Evaluation metrics for the Polynomial !!!
# the best-fitting polynomial will have lowest MSE (lower limit of 0), max R2 (upper limit 0f 1)
y_pred = p(x)

error = y_pred - y
MSE = (error**2).mean()
MSE
```

```python
SSR = (error**2).sum()   # sum of squares of Residuals / Errors
SST = ((y - y.mean())**2).sum()  # sum of squares of deviations of y from its mean
R2 = 1-(SSR/SST)
R2
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2_score(y, y_pred), mean_squared_error(y, y_pred)
```

```python
# for poly 3: MSE = 0.007001555965145752, R2 = 0.8703961578197009
# for linear-fit>> MSE = 0.053705239339019196, R2 = 0.005877350949013427 
```

```python
RMSE = np.sqrt(0.007001555965145752)   # Root MSE  >> sq. root of the MSE 
RMSE
```

```python
x = np.linspace(0,10, 100)
y = 2*x - 5 + np.random.rand(100)

fig = plt.figure(figsize = (8,5))  # this is to create a plot Axis (siize is provided as tuple)

plt.plot(x,y, 'bo', label='Original Data')

plt.xlabel('X values 0 to $\pi$', fontsize=18)
plt.ylabel('Y values ', fontsize=18)
plt.title('Curve Fitting', fontsize=21)
plt.legend(fontsize=15, loc = 'lower center')

# plt.xticks(xticks)
plt.grid(True)
plt.show()

# fig.savefig('name of the figure.png', dpi=200)
```

```python
MSE_val =[]
R2_val = []
for deg in range(1,15):
    p = np.poly1d(np.polyfit(x,y,deg)) 
    MSE_val.append(mean_squared_error(y, p(x)))
    R2_val.append(100*r2_score(y, p(x)))
```

```python
degrees = list(range(1,15))

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(degrees, MSE_val, 'b-o')
plt.ylabel('MSE Values')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(degrees, R2_val, 'r-o')
plt.ylabel('R2 Values')
plt.grid(True)
```
