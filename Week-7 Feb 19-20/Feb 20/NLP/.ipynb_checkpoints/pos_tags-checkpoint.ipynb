```python
import re
import pandas as pd
pd.set_option('display.max_colwidth', 300)

# load twitter data
tweets = pd.read_csv("tweets.csv")
```

```python
def cleaner(x):
    x = re.sub(r'http\S+', '', x) # remove URLs
    x = re.sub(r'[%$#@&)(]', '', x) # remove special characters
    x = re.sub(r'[.,:;]', '', x)
    return x    
```

```python
tweets.head()
```

```python
tweets['clean_tweet'] = tweets['tweet'].apply(lambda x: cleaner(x))
```
